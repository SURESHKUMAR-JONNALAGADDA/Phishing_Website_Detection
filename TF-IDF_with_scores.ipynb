```python
import pandas as pd
df = pd.read_csv(r"datathon_train.csv", header=None,names=['url','label'])
df
```

```python
df.drop(df.index[100860:125898], inplace=True)
df
```

```python
for i in range(df.shape[0]):
    df['url'][i].lower()
```

```python
df['url']=df['url'].apply(lambda x:x.replace('://',' '))
df['url']=df['url'].apply(lambda x:x.replace('.',' '))
df['url']=df['url'].apply(lambda x:x.replace('/',' '))
df['url']=df['url'].apply(lambda x:x.replace('%',' '))
df['url']=df['url'].apply(lambda x:x.replace(';',' '))
df['url']=df['url'].apply(lambda x:x.replace('-',' '))
df['url']=df['url'].apply(lambda x:x.replace('=',' '))
df['url']=df['url'].apply(lambda x:x.replace('&',' '))
df['url']=df['url'].apply(lambda x:x.replace('_',' '))
df['url']=df['url'].apply(lambda x:x.replace('?',' '))
```

```python
df['url']=df['url'].apply(lambda x:x.split(' '))
```

```python
df
```

```python
df['url']=df['url'].apply(lambda x:' '.join(x))
```

```python
tfidf = TfidfVectorizer(sublinear_tf= True, 
                       norm= 'l2', 
                       ngram_range= (1,2), 
                       stop_words ='english')
feature1 = tfidf.fit_transform(df)
array_of_feature = feature1.toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['url'], df['label'],test_size=0.15, random_state = 0)
X_train_counts = tfidf.fit_transform(X_train)
```

```python
clf = MultinomialNB().fit(X_train_counts, y_train)
y_pred = clf.predict(tfidf.transform(X_test))
```

```python
print('\n Accuracy of the classifier is',metrics.accuracy_score(y_test,y_pred))
print('\n Confusion matrix')
print(metrics.confusion_matrix(y_test,y_pred))
print(matthews_corrcoef(y_test,y_pred))
```
